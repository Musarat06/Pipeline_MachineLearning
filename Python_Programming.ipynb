```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

from sklearn.tree import DecisionTreeClassifier

import warnings
warnings.filterwarnings("ignore") 
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
df.head()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df = df.drop(['Unnamed: 32'], axis = 1)
```

## Whether normalizaiton needed or not.

Logistic regression usually assum the data is normally distributed but if it not we need to normalize data using StandardScaler (if we have negative values in data) or MinMaxScaler (If no negative values are there). To check it whether the data are in same scale the best way is to find the mean and standard deviation of each feature and compare them. If they are very different from each other then surelly we need normalization. otherwise we don't need it. 

```python
df.describe()
```

The means are quite different form each so the data needs normalization.

```python
df.dropna()
```

```python
X = df.iloc[:,2:]
Y = df.iloc[:,1]
```

```python
Y
```

```python
# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X,Y, random_state=42)
```

```python
# create a scaler and fit it to the training data
scaler = StandardScaler()
scaler.fit(X_train)

# transform the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# train a logistic regression model on the scaled data
lr = LogisticRegression()
lr.fit(X_train_scaled, y_train)

# evaluate the model on the testing data
print(lr.score(X_test_scaled, y_test))
```

### Applying the same modle with Pipline

```python
# create an ML pipeline with scaling and logistic regression
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', LogisticRegression())
])

# fit the pipeline to the training data
pipe.fit(X_train, y_train)

# evaluate the pipeline on the testing data
score = pipe.score(X_test, y_test)
print(f"Accuracy: {score:.3f}")
```

### Pipline for Decision Tree

```python
# create an ML pipeline with a decision tree classifier
pipe = Pipeline([
    ('classifier', DecisionTreeClassifier())
])

# fit the pipeline to the training data
pipe.fit(X_train, y_train)

# evaluate the pipeline on the testing data
score = pipe.score(X_test, y_test)
print(f"Accuracy: {score:.3f}")
```
